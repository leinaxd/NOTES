Matplotlib tests & Synthetic data

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = np.array([np.arange(0,256)], dtype=np.uint8)
ones = np.ones((1, 256))
img = img.T.dot(ones)
```

```python
plt.imshow(img, cmap='bone')
plt.ylabel('img value');
```

You don't see as a linear map. because your vision system is logarithmic.

```python
plt.imshow(img.T, cmap='bone')
zeros = np.zeros_like(img, dtype=np.uint8)
new_img = np.stack([img, zeros, img.T]).transpose(1, 2, 0)
new_img = np.uint8(new_img)
print(new_img.shape)
plt.figure()
plt.imshow(new_img, cmap='bone')
```

You can see the combination of the red and blue channel for the ramp intensity as shown earlier.

Red channel goes from no red, to full red.
Blue channel goes from no blue to full blue.
