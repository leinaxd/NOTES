Matplotlib tests & Synthetic data

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = np.array([np.arange(0,256)], dtype=np.uint8)
ones = np.ones((1, 256))
img_1 = img.T.dot(ones)
img_2 = img_1.T
img_3 = np.zeros_like(img_1, dtype=np.uint8)
```

```python
plt.imshow(img_1, cmap='bone')
plt.ylabel('img value');
plt.figure()
plt.imshow(img_2, cmap='bone')

new_img = np.stack([img_1, img_3, img_2]).transpose(1, 2, 0)
new_img = np.uint8(new_img)
plt.figure()
plt.imshow(new_img, cmap='bone')
```

You don't see as a linear map. because your vision system is logarithmic.

You can see the combination of the red and blue channel for the ramp intensity as shown earlier.

Red channel goes from no red, to full red.
Blue channel goes from no blue to full blue.


# operations

You cannot do the same operations in uint8 as you did in float

```python
img_1 = np.uint8(img_1)
img_2 = np.uint8(img_2)
avg_1 = (img_1 + img_2)/2
avg_2 = img_1/2 + img_2/2

plt.imshow(avg_1, cmap='gray')
plt.figure()
plt.imshow(avg_2, cmap='gray');
```

But you can operate into a continuous double world and go back 

```python
img_1 = np.double(img_1)
img_2 = np.double(img_2)
avg_1 = (img_1 + img_2)/2
avg_2 = np.uint8(avg_2)
plt.imshow(avg_1, cmap='gray')
plt.figure()
plt.imshow(avg_2, cmap='gray');
```

Sum and mul

```python
[x, y] = np.meshgrid(np.arange(500), np.arange(500))
fig, ax = plt.subplots(2,2, figsize=(10, 10))
ax[0,0].imshow(x+y, cmap='gray')
ax[0,0].set_title('x+y')
ax[0,0].axis('off');
ax[0,1].imshow(x, cmap='gray')
ax[0,1].set_title('x')
ax[0,1].axis('off')
ax[1,0].imshow(y, cmap='gray')
ax[1,0].set_title('y')
ax[1,0].axis('off');
ax[1,1].imshow(x*y, cmap='gray')
ax[1,1].set_title('x*y')
ax[1,1].axis('off');
```

Discretize a Gaussian

```python
[x, y] = np.meshgrid(np.arange(500), np.arange(500))
center_x, center_y= 250, 250
sigma = 100
fn = 255*np.exp(-0.5*((x-center_x)**2 + (y-center_y)**2)/sigma**2)
fn = np.uint8(fn)
fig, ax = plt.subplots(2,4, figsize=(15, 10))
ax[0,0].imshow(fn, cmap='gray')
ax[0,0].set_title('G 256 values')
ax[0,0].axis('off');

fn = np.uint8(fn/2)
ax[0,1].imshow(fn, cmap='gray')
ax[0,1].set_title('G 128 values')
ax[0,1].axis('off')

fn = np.uint8(fn/2)
ax[0,2].imshow(fn, cmap='gray')
ax[0,2].set_title('G 64 Values')
ax[0,2].axis('off');

fn = np.uint8(fn/2)
ax[0,3].imshow(fn, cmap='gray')
ax[0,3].set_title('G 32 Values')
ax[0,3].axis('off');

fn = np.uint8(fn/2)
ax[1,0].imshow(fn, cmap='gray')
ax[1,0].set_title('G 16 Values')
ax[1,0].axis('off');

fn = np.uint8(fn/2)
ax[1,1].imshow(fn, cmap='gray')
ax[1,1].set_title('G 8 Values')
ax[1,1].axis('off');

fn = np.uint8(fn/2)
ax[1,2].imshow(fn, cmap='gray')
ax[1,2].set_title('G 4 Values')
ax[1,2].axis('off');

fn = np.uint8(fn/2)
ax[1,3].imshow(fn, cmap='gray')
ax[1,3].set_title('G 2 Values')
ax[1,3].axis('off');
```
