```python
import matplotlib.pyplot as plt
import numpy as np
import urllib.request
```

```python
url = "https://github.com/leinaxd/NOTES/raw/main/digital_image_processing/Lecture_3/bank.jpeg"
with urllib.request.urlopen(url) as url:
    img = plt.imread(url, format='jpg')
plt.imshow(img)
plt.axis('off')

print(img.shape)
```

```python
r_mask = img[:, :, 0]
g_mask = img[:, :, 1]
b_mask = img[:, :, 2]
plt.figure()
plt.imshow(r_mask, 'gray')
plt.title('red')
plt.axis('off')
plt.figure()
plt.imshow(g_mask, 'gray')
plt.title('green')
plt.axis('off')
plt.figure()
plt.imshow(b_mask, 'gray')
plt.title('blue')
plt.axis('off');
print('split into channels')
```
